**VECTOR OPERATIONS**

```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#setting seaborn style to default
sns.set() 
```

```python
#plotting a vector
plt.quiver(0,0,4,5)
plt.show() #showing the plot
```

```python
plt.quiver(0,0,4,5, scale_units='xy', angles='xy', scale=1, color='r') 
plt.quiver(0,0,-4,2, scale_units='xy', angles='xy', scale=1, color='b') 
plt.xlim(-8,8)
plt.ylim(-8,8)
```
