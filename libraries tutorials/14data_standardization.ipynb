DATA STANDARDIZATION  
The process of standardizing a data to common shapes and format  

```python
import numpy as np
import sklearn.datasets
from sklearn.preprocessing import StandardScaler # for scaling the data
from sklearn.model_selection import train_test_split # for splitting the data and training
import pandas as pd
```

```python
#loading a dataset
data = sklearn.datasets.load_breast_cancer()
print(data)
```

```python
#converting the data into pandas dataframe
df = pd.DataFrame(data.data, columns=data.feature_names)
df.head( )
```

```python
#no of rows and columns
print(df.shape)
```

```python
#checking for missing values
print(df.isna().sum())
```

```python
#Standardizing the data using StandardScaler
X = df #this is features
Y = data.target #this is target variable by analyzing features we need to predict this target variable

print(X)
print(Y) #0 is malignant and 1 is benign ; no need to standardize this
```

test size - is how much data we want to keep for testing -> 20% of data is kept for testing  
random_state - is used to keep the data same for every run ; 3 means random state is 3 ; if we keep 3 then data will be same for every run   
creating 4 arrays X_train, X_test, Y_train, Y_test  
we can also standardize before splitting the data but it is better to standardize after splitting the data  
but if our data has some outliers then we should standardize before splitting the data  
outliers are abnormal data  
for e.g.  mean smoothness has values between 0.05263 and 0.16340 but if there is a value 1000 then it is an outlier

```python
#before standardizing we need to split data into training data and test data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=3)
```

```python
print(X.shape, X_train.shape, X_test.shape)
print(Y.shape, Y_train.shape, Y_test.shape)
```

STANDARDIZE THE DATA

```python
#print the standard deviation of the data
#if std is 1 then the data is standardized ; not in same range
print(np.std(data.data))

#another way 
print(data.data.std())
```

```python
#load the standard scaler in scaler
scaler = StandardScaler()
```

```python
#fit the training data
scaler.fit(X_train)
```

```python
#transform the training data
X_train_standardized = scaler.transform(X_train)
```

We need to standardize X_test data but no need to fit it because X_test data value is small and we should use SatandardScaler of X_train for X_test

```python
X_test_standardized = scaler.transform(X_test)
```

```python
#Check the standard deviation of the standardized data
print(X_test_standardized.std())
print(X_train_standardized.std())
```
