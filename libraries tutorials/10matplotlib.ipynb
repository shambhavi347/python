#MATPLOTLIB
#useful for making plots

```python
#import matpoltlib
import matplotlib.pyplot as plt  #pyplot -is python plot
```

```python
#import numpy lib
import numpy as np

x = np.linspace(0,10,100) #100 points from 0 to 10
y = np.sin(x) #y = sin(x)
z = np.cos(x) #z = cos(x)

#print(x)
#print(y)
#print(z)

#plot x and y - get sin wave
plt.plot(x,y)

#plot x and z - get cos wave
plt.plot(x,z)

plt.show() #show the plot

```

```python
#Adding label to the plot
plt.plot(x,y,label='sin(x)')
plt.plot(x,z,label='cos(x)')

plt.xlabel('angle') #x-axis label
plt.ylabel('value') #y-axis label

plt.title('sin(x) and cos(x)') #title of the plot

plt.legend() #show the legend

plt.show() #show the plot
```

```python
#Parabaola

x = np.linspace(-10,10,20) #20 points from -10 to 10
y = x**2 #y = x^2

plt.plot(x,y,label='x^2')

plt.xlabel('x') #x-axis label
plt.ylabel('y') #y-axis label

plt.title('Parabola') #title of the plot
```

```python
#Plot it with different symbols

x = np.linspace(-10,10,20) #20 points from -10 to 10
y = x**2 #y = x^2

plt.plot(x,y,'r+') #red plus


```

```python
plt.plot(x,y,'g.') #green dot
```

```python
plt.plot(x,y,'b*') #blue star
```

```python
# multiple lines in the same plot

x = np.linspace(-5,5,50) #20 points from -5 to 5
plt.plot(x,np.sin(x),'g-',label='sin(x)')
plt.plot(x,np.cos(x),'b--',label='cos(x)')

plt.legend() #show the legend
```

```python
#Bar Graph
fig = plt.figure()
ax = fig.add_axes([0,0,1,1]) #left, bottom, width, height
lang = ['English','French','Spanish','Latin','German']
people = [100,50,150,40,70]

ax.bar(lang,people)
plt.xlabel('Language')
plt.ylabel('Number of People')

plt.title('Language vs People')

plt.show() #show the plot

```

```python
#PIE Chart

fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1]) #left, bottom, width, height

lang = ['English','French','Spanish','Latin','German']
people = [100,50,150,40,70]

ax.pie(people,labels=lang,autopct='%1.1f%%') #autopct - show percentage; how many decimal points we need

```

```python
#Scatter Plot

x = np.linspace(0,10,30) #30 points from 0 to 10

plt.scatter(x,np.sin(x),color='y',label='sin(x)')
plt.scatter(x,np.cos(x),color='g',label='cos(x)')

plt.xlabel('Angles') #x-axis label
plt.ylabel('Sine/Cosine values') #y-axis label

plt.title('Scatter Plot') #title of the plot
plt.legend() #show the legend

plt.show() #show the plot
```

```python
#3D Scatter Plot

fig2 = plt.figure()
ax = plt.axes(projection='3d')

z = 20*np.random.random(1000)
x = np.sin(z)
y = np.cos(z)

ax.scatter(x,y,z,c=z,cmap='Blues')
```
